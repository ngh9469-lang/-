```python
selector = ".book-subject"
elements = driver.find_elements(By.CSS_SELECTOR, selector)
elements[0].text
```

```python
#전체 책 제목 추출
title_list = []

for title in elements:
    title_list.append(title.text)
    
print(len(title_list))
title_list
```

### 책 url 수집

```python
# 첫번째 책 url 확인
elements[0].get_attribute('href')
```

```python
#전체 수집 후 리스트 저장

url_list = []

for element in elements:
    url = element.get_attribute('href')
    url_list.append(url)

print(len(url_list))
url_list
```

```python
df = pd.DataFrame({'title':title_list,'url':url_list})
df
```

```python
df.to_csv('위키독스.csv', encoding ='utf-8-sig', index=False)
```

```python
#데이터프레임, csv 저장
#통계 - 추천수 평균 구하기
#추천수 판매가 시각화 히스토그램, 상자그림
```

```python
# 블로그 링크(url) 불러오기
df_url = pd.read_csv("위키독스.csv")

print(len(df_url))
df_url.head(3)
```

```python
df_url['url']
```

```python
url_cnt = len(df_url['url'])
url_cnt
```

```python
# for문으로 전체 블로그 글 크롤링
chromedriver_autoinstaller.install(cwd=True)

total_dict = {}  # 전체 블로그 글 수집할 딕셔너리 선언 

# i = 0   # 첫번째 글 
for i in range(0, 20):   # blog_cnt     
    try:    
        driver = webdriver.Chrome()   # driver : 크롬 브라우저
        driver.get(df_url['url'][i])  # i가 바뀌면서 블로그 글도 계속 바뀐다.
        time.sleep(1)
        
        # iframe 접근(네이버 사이트들의 특성 - 네이버 블로그, 네이버 카페)
        #driver.switch_to.frame('mainFrame')
        
        # 한 개 블로그 글(제목, 글쓴이, 날짜, 내용 등) 크롤링 후 저장할 딕셔너리
        target_info = {}    # 선언, 초기화 
        
        # 블로그 글 제목 첫번째 크롤링
        selector = "#load_content > div.book-info.row.clearfix > div:nth-child(2) > div:nth-child(1) > a"    
        title_raw = driver.find_element(By.CSS_SELECTOR, selector)          # title
        title = title_raw.text
        target_info['title'] = title
        
        # 편집일시
        selector = "#load_content > div.book-info.row.clearfix > div:nth-child(2) > div:nth-child(2)"
        nickname = driver.find_element(By.CSS_SELECTOR, selector).text
        target_info['nickname'] = nickname   # 찾은 닉네임 데이터를 target_info에 저장 
        
        # 판매가
        selector = "#load_content > div.book-info.row.clearfix > div:nth-child(2) > div:nth-child(4)"
        price = driver.find_element(By.CSS_SELECTOR, selector).text
        # target_info 딕셔너리에 날짜 저장
        target_info['price'] = price


        
        # 반복문으로 모든 문단 텍스트 수집 후 리스트로 저장
        overview_str = ''
        selector = "#load_content > div.page-content > p"
        elements = driver.find_elements(By.CSS_SELECTOR, selector)
for element in elements:
            
            overview_str = overview_str + element.text + '\n'    # '\n' : 줄바꿈(엔터)
        #for review in text_raw:
        #    review_str = review_str + review.text.replace("\n", " ")
            # print(review.text)
            target_info['element'] = overview_str
    
        print(target_info)
        total_dict[i] = target_info   # i가 0~119까지 변하면서 각 블로그 글을 저장.
        print(i+1, "번째 데이터가 수집되었습니다.", "\n")

        time.sleep(2)

        driver.close()      # 블로그 글 한 개 마다 크롤링했던 브라우저 종료
        
    except:
        print(i+1, "번째 블로그는 건너 뜁니다.")
        driver.close()      # 크롬 브라우저 종료
        continue
```
