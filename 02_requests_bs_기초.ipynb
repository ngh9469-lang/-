```python
from bs4 import BeautifulSoup as bs
```

```python
# BeautifulSoup이란 : 파이썬 라이브러리로, HTML 문서를 파싱하고 분석하는 데 사용되는 도구
# 파싱이란? 문서에 있는 구성 요소를 분해하고 추출하는 것
```

```python
html_doc1 = """
<!doctype html>
<html>
    <head> 
        <title> 기초 크롤링 </title> 
    </head>
    <body> 
        크롤링을 해봅시다. 
    </body>
</html>
"""
```

```python
# bs(BeuatifulSoup)의 역할 : html 문서를 정리해준다.
soup = bs(html_doc1, 'html.parser')
soup
```

```python
# title 데이터 크롤링  #방법1
soup.select_one('title').text.strip()
```

```python
# title 데이터 크롤링  #방법2
soup.head.title.text.strip()
```

```python
# title 데이터 크롤링  #방법3
soup.find('head').title.text.strip()
```

```python
#body 데이터 수집
soup.find('body').text.strip()
```

# 실습2

```python
html_doc2 = """
<!doctype html>
<html>
    <head> 
        기초 크롤링 따라하기 
    </head>
    <body> 
        <div> 첫 번째 부분 </div>
        <div> 두 번째 부분 </div>
    </body>
</html>
"""
```

```python
soup2 = bs(html_doc2, 'html.parser')
soup2
```

```python
# 1. head 안의 텍스트 추출
soup2.find('head').text.strip()
```

```python
# 2. 모든 div 찾고 리스트로 저장
# find : 첫번째 요소 찾기
# find_all : 모든 요소 찾고 리스트로 저장.
soup2.find_all('div')
```

```python
# 3. 찾은 div 중에서 2번째 부분 인덱싱하고 텍스트 추출
soup2.find_all('div')[1].text.strip()
```
