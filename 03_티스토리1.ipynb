```python
from urllib.request import urlopen    # urlopen : URL에 접속해서 정보를 받아오는 함수  requests.get 그냥써라  urlopen말고
from bs4 import BeautifulSoup as bs   # BeautifulSoup: HTML 문서를 정제하는 라이브러리  
#bs4 상자 안에서 BeautifulSoup만 꺼내서 바로 쓰겠다
import requests


```

```python
url = 'https://ai-dev.tistory.com/1'
html_encoded = requests.get(url).text
soup = bs(html_encoded, 'html.parser')
soup
```

```python
soup = bs(html_encoded, 'html.parser')
soup
```

```python
#글 제목 찾기-> h1 태그

soup.find_all('h1')
```

```python
title = soup.find_all('h1')[1].text.strip()
title
```

```python
# 블로그 제목 찾기   방법2           #간단한 사이트에서 찾기 쉬움
soup.find_all('h1')[0].text
```

```python
# 블로그 제목 id로 찾기   방법2   
soup.find('header', id='header').find('h1').text     #'태그', id='아이디명'  #좁잡한 사이트느 h1이 많아 아이디로 찾는게 좋다
#soup.find('태그', id='아이디명').find('h1').text.strip()
```
