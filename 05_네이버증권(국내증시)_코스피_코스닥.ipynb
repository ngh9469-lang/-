```python
<실습 - 네이버증권 - 국내증시 https://finance.naver.com/sise/>
# 코스피, 코스닥, 코스피200 지수 크롤링하기
# 인기 검색 종목 크롤링하고 데이터프레임으로 만들기
# Top종목, 상한가 -> 데이터프레임으로 변환 후 csv로 저장


import requests
from bs4 import BeautifulSoup as bs

import pandas as pd
import numpy as np
```

```python
# requests.get() : 웹 페이지에서 html 정보를 받아옴 
response = requests.get("https://finance.naver.com/sise")  
response
```

```python
html = response.text

soup = bs(html, 'html.parser')
soup
```

# 코스피 지수, 등락률

```python
# 코스피 지수
kospi = soup.find(id="KOSPI_now").text
kospi
```

```python
# 실수화
if ',' in kospi:
    kospi = float(kospi.replace(",", ""))
else:
    kospi = float(kospi)
kospi
```

```python
# 등락값
kospi_change = soup.find(id="KOSPI_change").text
ksopi_change_list = kospi_change.strip().split(" ")
ksopi_change_list
```

```python
kospi_change = float(ksopi_change_list[0])
kospi_change
```

```python
# 등락률
kospi_change_pct = ksopi_change_list[1]
kospi_change_pct = kospi_change_pct.split("%")[0] + "%"
kospi_change_pct
```

```python
# 수집한 kospi 데이터를 딕셔너리 형태로 저장하기.
kospi_dict = {}                  # 빈 딕셔너리를 하나 만든다.
kospi_dict['값'] = kospi     # '코스피'라는 key값을 추가. value는 kospi값 
kospi_dict['등락값'] = kospi_change
kospi_dict['등락률'] = kospi_change_pct

kospi_dict
```

```python
# 문자와 숫자가 붙어있는 경우 분리하려면?
text = "삼성전자160,900"
text
```

```python
import re
```

```python
#1. 글자 한글 영어 등 만 찾기 : \D+ (숫자가 아닌 것들이 연속됨 소대문자 의미 다름)
name = re.findall(r'\D+', text)
name
```

```python
#1. 글자 한글 영어 등 만 찾기 : \D+ (숫자가 아닌 것들이 연속됨 소대문자 의미 다름) 
name = re.findall(r'\D+', text)[0]
name
```

```python
# 2. 숫자와 쉼표만 찾기: [0-9,]+ (0~9까지 숫자와 쉼표가 연속됨)
price = re.findall(r'[0-9,]+', text)[0]

price = int(price.replace(",", ""))
price
```
