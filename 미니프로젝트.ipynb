```python
# 형태소 분석기 초기화
okt = Okt()

# 1. 명사 추출 및 2글자 이상인 단어만 필터링
nouns = okt.nouns(total_str)
words = [n for n in nouns if len(n) > 1]

# 2. 단어 빈도 계산
count = Counter(words)
top_10 = count.most_common(10)

print("----- 빈출 단어 TOP 10 -----")
for word, freq in top_10:
    print(f"{word}: {freq}회")
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정 (Windows용)
rc('font', family='Malgun Gothic')

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 빈출 단어 데이터
words = ['프로', '단백질', '음료', '추천', '랩노쉬', '링크', '제로', '슈가', '틴드', '후기']
counts = [33, 31, 30, 25, 18, 17, 12, 12, 11, 11]

# 색상 설정 (랩노쉬만 강조)
colors = ['orange' if word == '랩노쉬' else 'skyblue' for word in words]

# 그래프 생성
plt.figure(figsize=(10, 6))
bars = plt.bar(words, counts, color=colors)

# 값 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5,
             f'{int(height)}', ha='center', va='bottom')

# 제목 및 라벨
plt.title('프로틴 단백질 음료 추천 키워드 빈출 단어 TOP 10', fontsize=14)
plt.xlabel('단어')
plt.ylabel('언급 횟수')

# 가독성 개선
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
